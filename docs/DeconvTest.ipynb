To evaluate the performance of various deconvolution methods, we need to compare
the deconvolved image to a ground truth, which is readily provided by
synthetic data. Such synthetic data can be generated with our DeconvTest framework: a python-based simulation framework that allows the user to quantify and compare the 
performance of different deconvolution methods. The framework integrates all components needed for such 
quantitative comparison: it (1) generates synthetic ground truth images through in silico microscopy experiments, 
(2) deconvolves the images with multiple algorithms, and (3) quantifies the deconvolution accuracy by various 
performance measures.

# Simulation of microscopy experiments 

Microscopic imaging of a sample involves illuminating the sample with a light source, collecting
the emitted light, and discretizing the signal by recoding it in a detector, such as a CCD camera. This process can
be represented as a continuous convolution of the sample with a point spread function (PSF) -- the image of a point source -- and subsequent discretization. Convolution of complex continuous signals is non-trivial to compute.
Hence, when simulating this process on a computer, we represent the continuous convolution process
by a convolution of discrete signals of high resolution (very small voxel size), while the process of discretization is simulated by the subsequent downsampling (decrease of the resolution).

￼<img src="img/in_silico_microscopy.png" width="800">

To simulate the whole microscopy process, we need to model the following steps and components:
1. Images of cells
2. Point spread function (PSF) of the imaging system
3. Convolution and downsampling
4. Noise

## Adding synthetic noise

To investigate the influence of noise, Poisson or Gaussian noise of different signal-to-noise ratios (SNRs) can be added to the convolved and down-sampled images. 


The SNR for the Gaussian noise is evaluated as follows:

$$\mathrm{SNR} = 20 \log_{10} \frac{I_{max}}{\sigma}$$

Here, $I_{max}$ is the maximum image intensity, which represents the signal amplitude; $\sigma$ is the standard deviation of the Gaussian noise, which represents the noise amplitude.

Hence, to add Gaussian noise of a given SNR, the noise amplitude $\sigma$ can be computed as follows: 
$$\sigma = \frac{I_{max}}{10^{\mathrm{SNR}/20}}$$

Then, for each pixel, a random number is drawn from the normal distribution with $\mu = 0$ and $\sigma = \sigma$ and is added to the value of this pixel.


The SNR of the Poisson distribution is evaluated as follows:

$$\mathrm{SNR} = \sqrt{I},$$

To generate the Poisson noise with a given SNR, the image is normalized such that its maximum value equals to $\mathrm{SNR}^2$. Then, each pixel's intensity is replaced by a random value from the Poisson distribution that has the expectation value equal to corresponding pixel intensity.


### Usage examples
##### Example 1
Add Poisson noise with SNR = 2 to the downsampled cells from [Example 1 of section 2.5.1](#downsizing-example1):

```python
cell.add_poisson_noise(snr=2)
cell.show_2d_projections()
```

##### Example 2
Generate a spherical cell of size 10 µm and voxel size 0.3 µm per px and add Gaussian noise with SNR = 5:

```python
cell = Cell(resolution=0.3, size=10)
cell.add_gaussian_noise(snr=5)
cell.show_2d_projections()
```

##### Example 3
Generate a multicellular sample with 6 ellipoidal cells with axes sizes 7 $\pm$ 2 µm; sample dimensions 50 x 50 x 50 µm and voxel size 0.3 µm per px, and add Gaussian noise with SNR = 2:

```python
params = CellParams(number_of_cells=6, size_mean_and_std=(7, 2), 
                    equal_dimensions=False)
stack = Stack(cell_params=params, resolution=0.3, stack_size=[50, 50, 50])
stack.add_gaussian_noise(snr=2)
stack.show_2d_projections()
```

## Iterative Deconvolve 3D (DAMAS) 

DAMAS is an iterative deconvolution algorithm that includes multiple regularization techniques (see Dougherty *et al.* 2005 and the plugin documentation at <http://www.optinav.info/Iterative-Deconvolve-3D.htm> for details). 

Six of the plugin's parameters can be adjusted from the DeconvTest framework:
- Normalize PSF (normalizes the PSF before deconvoltution)
- Detect divergence (stops the iteration if the changes in the image increase)
- Perform antiringing step (reduces artifacts from features close to the image border)
- Wiener filter $\gamma$ (regualization parameter)
- Low pass filter (regualization parameter)
- Terminate iteration if mean delta less than x% (stops the iteration if the changes in the image are less than the value of x)

￼<img src="img/damas.png" width="600">

R. Dougherty, “Extensions of DAMAS and Benefits and Limitations of Deconvolution in Beamforming,”
(American Institute of Aeronautics and Astronautics, 2005).

- Detect divergence: False
- Perform antiringing step: True
- Wiener filter $\gamma$: 0.0001
- Low pass filter: 3
- Terminate iteration if mean $\delta$ <: 0.01

# Running deconvolution experiments with the "run_experiment" script

## Specifying parameters of the experiment 

An *in silico* microscopy and deconvolution experiment can be run via executing a specifically developed script, which accepts the parameters of the experiment specified in a csv file.

The following parameters are used by the script:

- **simulation_folder**: directory to save the simulation output
- **max_threads**: the maximal number of processes to run in parallel
- **print_progress**: if True, the progress of the computation will be printed
- **cell_parameter_filename**: csv file name in the `simulation_folder` to save cell parameters 
- **cell_parameter_folder**: directory in the `simulation_folder` to save cell parameters for mutiple multicellular stacks
- **number_of_stacks**: number of multicellular stacks to generate; if None, no stack will be generated, but only single cells
- **number_of_cells**: number of single cells to generate or number of cells per each multicellular stack
- **size_mean_and_std**: mean value and standard deviation for the cell size in micrometers; the cell size is drawn randomly from a Gaussian distribution with the given mean and standard deviation.
- **equal_dimensions**: if True, input cells will be generated as spheres; if False, input cells will be generated as ellipsoids with sizes for all three axes chosen independently
- **spikiness_range**: range for the fraction of cell surface area covered by spikes
- **spike_size_range**: range for the standard deviation for the spike amplitude relative to the cell radius
- **spike_smoothness_range**: range for the width of the Gaussian filter that is used to smooth the spikes
- **inputfolder**: directory in the `simulation_folder` to save generated cells or multicellular stacks
- **input_voxel_size**: voxel size in z, y and x used to generate the cell image; if one value is provided, the voxel size is assume to be equal along all axes.
- **stack_size_microns**: dimensions of the multicellular image stack in micrometers
- **psffolder**: directory in the `simulation_folder` to save generated PSF images
- **psf_sigmas**: standard deviations of the PSF in xy in micrometers
- **psf_elongations**: PSF aspect ratios
- **convolution_results_folder**: directory in the `simulation_folder` to save the results of convolution of input cells with PSFs
- **resizing_results_folder**: directory in the `simulation_folder` to save the downsized images
- **voxel_sizes_for_resizing**: list of new voxel sizes to which the input images should be resized; each item of the list is a scalar (for the same voxels size along all axes) or sequence of scalars (voxel size in z, y and x)
- **noise_results_folder**: directory in the `simulation_folder` to save images after adding noise
- **gaussian_snrs**: list of target signal-to-noise ratios (SNR) after adding the Gaussian noise; None corresponds to no Gaussian noise
- **poisson_snrs**: list of target signal-to-noise ratios (SNR) after adding the Poisson noise; None corresponds to no Poisson noise.
- **deconvolution_results_folder**: directory in the `simulation_folder` to save deconvolved images
- **algorithm**: deconvolution algorithm(s) that will be applied to deconvolve the input images; if a sequence of algorithms is provided, all algorithms from the sequence will be applied
- **rif_lambda**: regularization parameter for the RIF algorithm; if a sequence is provided, all values from the sequence will be tested
- **rltv_lambda**: regularization parameter for the RLTV algorithm; if a sequence is provided, all values from the sequence will be tested in combination with the `iterations` parameter
- **iterations**: number of iterations in the RLTV algorithm; if a sequence is provided, all values from the sequence will be tested in combination with the `rltv_lambda` parameter
- **normalize**: `Normalize PSF` parameter for the Iterative Deconvolve 3D plugin; if a sequence is provided, all values from the sequence will be tested in combination with other parameters
- **perform**: `Perform anti-ringig step` parameter for the Iterative Deconvolve 3D plugin; if a sequence is provided, all values from the sequence will be tested in combination with other parameters
- **detect**: `Detect divergence` parameter for the Iterative Deconvolve 3D plugin; if a sequence is provided, all values from the sequence will be tested in combination with other parameters
- **wiener**: `Wiener filter gamma` parameter for the Iterative Deconvolve 3D plugin; <0.0001 to turn off, 0.0001 - 0.1 as tests; if a sequence is provided, all values from the sequence will be tested in combination with other parameters
- **low**: `Low pass filter` parameter for the Iterative Deconvolve 3D plugin, pixels; the same value is used for low pass filter in xy and z; if a sequence is provided, all values from the sequence will be tested in combination with other parameters
- **terminate**: `Terminate iteration if mean delta < x%` parameter for the Iterative Deconvolve 3D plugin; 0 to turn off; if a sequence is provided, all values from the sequence will be tested in combination with other parameters
- **log_computing_time**: if True, computing time spent on deconvolution will be recorded and stored in the `log_folder`
- **log_folder**: directory in the `simulation_folder` to store computing time
- **segmentation_results_folder**: directory in the `simulation_folder` to save segmentation results
- **preprocess**: if True, the image will be preprocessed with a median filter (size 3) prior to segmentation
- **thr**: threshold value for image segmentation; if None, automatic Otsu threshold will be computed
- **relative_thr**: if True, the value of `thr` is multiplied by the maximum intensity of the image
- **postprocess**: if True, morphological opening and closing and binary holes filling will be applied after theresholding
- **accuracy_results_folder**: directory in the `simulation_folder` to save the deconvolution performance results


## Running the experiment with the "run_experiment" script 

To run a deconvolution experiment with specified settings, the `run_simulation.py` scipt should be copied to the directory of choice, and the following command should be executed:

`python run_simulation.py <settings_file>`

This command will execute all steps of the deconvolution experiment. To execute only one of the steps (generate synthetic data, perform deconvolution, or evaluate deconvolution performance), corresponding step number (1, 2, or 3) should be provided as an argument:

`python run_simulation.py <settings_file> <experiment_step>`,

where `<experiment_step>` is an integer 1 (for the generating synthetic data), 2 (for the deconvolution step), or 3 (for evaluating the deconvolution performance). 

For example, the following command will generate synthetic data with parameters specified in the `settings.csv` file:

`python run_simulation.py settings.csv 1`
