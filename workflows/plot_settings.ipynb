```python
from __future__ import print_function, unicode_literals, absolute_import, division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from skimage import io
```

```python
path = '../Data/qualitative_settings/'
name = 'cell_004.png'
inputpath = 'input_maxproj/'
convpath = 'noise_maxproj/psf_sigma_1.2_aspect_ratio_4.5_voxel_size_[0.3_0.3_0.3]_noise_poisson_snr=2.0/'
deconvpath = 'deconvolved_upsampled_maxproj/'
deconvpath2 = 'psf_sigma_1.2_aspect_ratio_4.5_voxel_size_[0.3_0.3_0.3]_noise_poisson_snr=2.0_voxel_size_[0.1_0.1_0.1]/'
comparepath = ['deconvolution_lab_rif_regularization_lambda=0.001/',
               'deconvolution_lab_rif_regularization_lambda=0.01/',
               'deconvolution_lab_rif_regularization_lambda=0.1/',
               'deconvolution_lab_rif_regularization_lambda=1.0/', 
               'deconvolution_lab_rif_regularization_lambda=10.0/', 
               'deconvolution_lab_rif_regularization_lambda=100.0/', 
               'deconvolution_lab_rif_regularization_lambda=1000.0/',
               'deconvolution_lab_rif_regularization_lambda=10000.0/',]
```

```python
stack1 = [io.imread(path + inputpath + name),
        io.imread(path + convpath + name)]
labels1 = ['input', 'convolved']
stack2 = []
labels2 = []
for cpath in comparepath:
    stack2.append(io.imread(path + deconvpath + cpath + deconvpath2 + name))
    labels2.append('$\lambda$ = ' + cpath.split('=')[-1].split('/')[0])
pad_width = int((stack2[0].shape[1] - stack1[0].shape[1])/2)
stack1[0] = np.pad(stack1[0], pad_width)
print(stack1[0].shape, stack2[0].shape)
```

```python
size = (7.5,12)
fig, axs = plt.subplots(2, 5, figsize=size)
for i in range(len(stack1)):
    plt.sca(axs[0,i])
    io.imshow(stack1[i], cmap='viridis')
    plt.axis('off')
    plt.title(labels1[i])
    
for i in range(3):
    plt.sca(axs[0, i+2])
    io.imshow(stack2[i], cmap='viridis')
    plt.axis('off')
    plt.title(labels2[i])
    
    
for i in range(len(stack2)-3):
    plt.sca(axs[1, i])
    io.imshow(stack2[i+3], cmap='viridis')
    plt.axis('off')
    plt.title(labels2[i+3])
plt.tight_layout()
plt.savefig('plot_settings.svg', dpi=300)
```
